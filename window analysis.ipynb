```python
def makeSums(workingDF):
    #aggregate area by time, so 2 windows at 7am with 1m² each would come up as 2m²
    windows = workingDF.groupby("time")
    sums = windows.aggregate(np.sum)
    sums["exposed"]        = sums["exposedArea"] > 0 #true if any exposed area 
    sums["oneSqm_Exposed"] = sums["exposedArea"] > 1 #true if exposed area is bigger than 1m²
    return sums
```

```python
def makeTimeDF(sums):
    #hack section to fix the indexing
    time = sums.index.tolist()
    area = sums["exposedArea"].tolist()
    data = np.zeros((len(sums),), dtype=[('time', 'i4'), ('exposedArea', 'f4')])
    data[:] = zip(time,area)
    timeDF = pd.DataFrame(data)
    return timeDF
```

```python
def assess(dfForOneWindow):
    """Takes a dataframe describing one window and marks it as 
    according to the ammount of sunlight each row recieves.
    
    *          exAny: any sun in this row, one pixel of light passes, at any time
    *          exGt1: more than 1m², at any time
    * exAnyInBracket: any sun in this row, one pixel of light passes, between 9am and 3pm
    * exGt1InBracket: more than 1m², between 9am and 3pm
    
    This means that any time outside the bracket will always fail the '...InBracket' 
    regardless of how much light it gets.
    
    returns a dictionary of the pass status.
    """
    exAny = dfForOneWindow["exposed"].tolist().count(True)
    exGt1 = dfForOneWindow["oneSqm_Exposed"].tolist().count(True)
    
    exAnyInBracket = dfForOneWindow[8:33]["exposed"].tolist().count(True)
    exGt1InBracket = dfForOneWindow[8:33]["oneSqm_Exposed"].tolist().count(True)
    
    return {"exAny":exAny,
            "exGt1":exGt1,
            "exAnyInBracket":exAnyInBracket,
            "exGt1InBracket":exGt1InBracket}
```

```python
assessedAppts = []
timeArea      = []
for name, group in apptGroups:
    #print "Building {} - Level {} - Appartment {}".format(*name)
    #remove the columns that aren't important
    workingDF = group.drop(["appartment", "appartment_window", "building", "building_level",
                            "faceID", "filename", "name", "path", "level", "appt",
                            "hour", "minute", "month", "pcWhite"],1).sort("time")
    #print workingDF
    
    sums = makeSums(workingDF)
    
    timeDF = makeTimeDF(sums)
    
    timeArea.append([timeDF, name])
    #print timeDF#[8:33]#8 is 9am, 33 is 3pm

    a=assess(sums)
    #print u"time slots exposed", a['exAny']
    #print u"time slots exposed > 1m²", a['exGt1']
    rtn = {"Building":      name[0],
           "Level":         name[1],
           "Appartment":    name[2],
           "exAny":         a['exAny'],
           "exGt1":         a['exGt1'],
           "exAnyInBracket":a["exAnyInBracket"],
           "exGt1InBracket":a["exGt1InBracket"]}
    #print rtn
    assessedAppts.append(rtn)

print "assessedAppts:\n", assessedAppts[:5]
print "timeArea:\n", timeArea[:5]
```

```python
def classifyAppt(x,bracketLength=8):
    print x
    if x["exGt1InBracket"] >= bracketLength: #8 is 2 hours in chunks of 15 minutes
        return (u"1) Hard Pass! 2h @ >1m & in ||" ,1)
    
    elif x["exAnyInBracket"] >= bracketLength:
        return (u"2) Soft Pass - 2h in || (old hard)" ,2)
    
    elif x["exGt1"] >= bracketLength:
        return (u"3) Soft Pass - 2h @ >1m (hard out of bracket))" ,3)
    
    elif x["exGt1"] > 0:
        return (u"4) Non Zero - >1m" ,4)
    
    elif x["exAny"] > 0:
        return (u"5) Non Zero" ,5)
    
    else:
        return ("6) fail",6)
```

```python
levelSummary = apptsDF.groupby(["Building","Level"])
statuses = sorted(apptsDF["status"].unique())
levelSummaryDicts = []
for name, group in levelSummary:
    name =  "Building {} Level {}".format(*name)
    stats = {"Level":name}
    for status in statuses:
        stats[status] = group["status"].tolist().count(status)
        #print group["status"].tolist().count(status),status
    levelSummaryDicts.append(stats)

levelDF = pd.DataFrame(levelSummaryDicts)
levelDF["numAppts"]   = levelDF.apply(lambda x: sum([x[0],x[1],x[2],x[3],x[4]]), axis=1)
levelDF["pcHardPass"] = levelDF.apply(lambda x: x[0]/float(x["numAppts"]), axis=1)
levelDF["pcSoftPass"] = levelDF.apply(lambda x: sum([x[1],x[2],x[3]])/float(x["numAppts"]), axis=1)
levelDF["pcfail"]     = levelDF.apply(lambda x: x[4]/float(x["numAppts"]), axis=1)
#levelDF["pcCheck"]     = levelDF.apply(lambda x: sum([x["pcHardPass"], x["pcSoftPass"], x["pcfail"]]), axis=1)
levelDF.to_csv("./LevelSummary.csv")
levelDF#.head()
```
